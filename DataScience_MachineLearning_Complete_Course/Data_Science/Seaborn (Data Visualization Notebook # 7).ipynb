### styles and fonts with Regression plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
tips.rename(columns = {'sex':'gender'}, inplace = True)
tips.head()
```

### setting style:

```python
sns.set_style('white')
sns.countplot(x = 'gender', data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x = 'gender', data = tips)
```

```python
sns.set_style('whitegrid') # sets the style to whitegrid
sns.countplot(tips['gender'], data = tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'gender', data = tips)
sns.despine(left = True, bottom = True) # removes the plot axes from left and bottom
```

```python
plt.figure(figsize = (8,3)) # setting figure size using matplotlib
sns.set_style('dark')
sns.distplot(tips['tip'], kde = True)
sns.despine(left = True, bottom = True)
```

```python
sns.set_context('poster') # sets the figure size for poster print
sns.set_style('darkgrid')
sns.rugplot(tips['total_bill'])
sns.despine(left = True, bottom = True)
```
