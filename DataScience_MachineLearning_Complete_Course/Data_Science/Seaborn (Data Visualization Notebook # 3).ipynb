### Seaborn Categorical Plots:

#### bear in mind the plots we are gonna study in this Notebook are mostly categorical plots, except the countplot. We pass in x as  a 'categorical value' and y as a 'numerical value' in order to see the visualization. Let's go ahead and play with some plots to visualize our data

```python
import numpy as np
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head() # loading the head of our dataset
```

### barplot():

#### categorical plot which is univarite and shows target column values categorically

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips) # shows total_bill column with respect to 
# categories in 'sex' column

# 'x' --> categorical column
# 'y' --> numerical column
```

```python
# we can also set the estimator function for our barplot
# estimator is the function which user can choose in order to see the desired results
# e.g lets check out standart deviation of our numerical column
# for this we can add estimator argument in our barplot

sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

### countplot():

```python
sns.countplot(tips['sex']) # shows the counts / category
```

```python
tips['sex'].value_counts() # results match up
```

### boxplot():

#### categorical bivariate plot takes in 'x' column as categorical and 'y' as numerical 

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')

# categorized by col 'day', hue 'smoker'
```

### violinplot():

#### categorical bivariate plot takes in 'x' column as categorical and 'y' as numerical

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')

# categorized with col 'day' and 'hue'
```

```python
sns.violinplot(x = 'sex', y = 'total_bill', data = tips, hue = 'smoker')

# seems like we have more smokers both in male and female customers :)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)

# splitting the values off of hue
```

### stripplot():

```python
sns.stripplot(x = 'sex', y = 'total_bill', data = tips)

# shows the data in strips for visualization
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.stripplot(x = 'sex', y = 'total_bill', data = tips, hue = 'day')

# categorizing on the basis of col 'sex', and hue 'day'
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)

# jitter: adds a bit more random noise in order to see the data point more clearly
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', jitter = True, split = True)

# jitter: adds a bit more random noise in order to seperate the data observations
# split: splits the graph strips on the basis of hue, so for male and female we have seperate strips
```

### swarmplot():

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)

# combination of scatter, strip and violin plots
```

```python
sns.swarmplot(x = 'sex', y = 'total_bill', data = tips, hue = 'day', split = True)
```

```python
# we can also call both 'violin' and 'swarm' plots in the same canvas
# and visualize some cool results

sns.violinplot(x = 'sex', y = 'total_bill', data = tips)
sns.swarmplot(x = 'sex', y = 'total_bill', data = tips, color = 'black')

# see how cool it looks :)
```

### factorplot():

#### This is the most general method of calling all the plots, by passing in the name of that specific plot
#### inside the 'kind' argument. Let's go ahead a give it a try

```python
sns.factorplot(x = 'sex', y = 'total_bill', data = tips, kind = 'bar')

# passing in argument 'kind' = barplot
```

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'strip')

# passing in kind argument = 'strip' draws a stripplot for us
```
