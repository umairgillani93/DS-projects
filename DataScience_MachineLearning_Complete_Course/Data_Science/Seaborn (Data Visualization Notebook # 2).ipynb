```python
# lets revise some of the plots we studied using seaborn in Notebook # 1 real quick :)

import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
# distribution plot

sns.distplot(data['total_bill'], kde = False)
```

```python
# jointplot

sns.jointplot(x = data['total_bill'], y = data['tip'], data = data)
```

```python
# pairplot

sns.pairplot(data, hue = 'sex')
```

```python
# rugplot

sns.rugplot(data['total_bill'])
```

```python
# kdeplot

sns.kdeplot(data['total_bill'])
```
