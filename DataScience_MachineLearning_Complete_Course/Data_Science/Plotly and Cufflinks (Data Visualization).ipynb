### Plotly and Cufflinks:

#### Another interesting library in Python, through which we can visualize our plots even better is 'Plotly'. Plotly uses yet another library called 'Cufflinks' in order to connect it with Pandas to visualize Pandas DataFrames. Lets go ahead and try to visualize our DataFrames with Plotly

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from plotly import __version__ # reports back the version of plotly
```

```python
print(__version__) # prints the version of plotly
```

```python
# lets go ahead and now import cufflinks in order to connect our plotly with pandas

import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

# since plotly is a company that broadcasts your data and make it able to visualize on their forums
# we need to import some methods 'offline', in order not to pay them using their survice
```

```python
# now in order to have everything work in notebook we need to apply the following commands

init_notebook_mode(connected = True) # this is going to connect plotly with our notebook{}
```

```python
# finally we need to use 'cufflinks' as offline as well
# so for this we are gonna call the following method off of our cufflinks library

cf.go_offline()
```

```python
df = pd.DataFrame(data = np.random.randn(100,4), columns = 'A B C D'.split())
```

```python
df.head()
```

```python
df1 = pd.DataFrame(data = {'Category':'A B C'.split(), 'Values':[32,43,50]})
```

```python
df1
```

```python
# if we just go ahead and call a simple plot off of our DataFrame
# we can jus call 'plot' method and I'll create plot using matplotlib

df.plot()
```

```python
# but what if we call 'iplot' method off of our DataFrame

df.iplot() # showing values of both rows and columns using iplot method from plotly
```

```python
df.iplot(x = 'A', y = 'B', kind = 'scatter') # kind specifies the type of plot we wanna create with plotly
```

```python
df.iplot(x = 'A', y = 'B', kind = 'scatter', mode = 'markers')

# we can also specify mode of our plot
```

```python
df.iplot(x = 'A', y = 'B', kind = 'scatter', mode = 'markers', size = 20)

# size can also be adjusted as well
```

```python
df1.iplot(x = 'Category', y = 'Values', kind = 'bar')

# creating bar plot off of plotly
```

```python
df.iplot(kind = 'bar')
```

```python
df.iplot(kind = 'box')

# we can draw different kinds of plots off of plotly
```

```python
df2 = pd.DataFrame({'x':[1,2,3,4,5], 'y':[10,20,30,40,50], 'z':[500,400,300,200,100]})
```

```python
df2
```

```python
df2.iplot(kind = 'surface')

# now this is create a 3-D surface plot
```

```python
df2.iplot(kind = 'surface', colorscale = 'rdylbu')

# we can also change colorscale as well
```

```python
df1.iplot(kind = 'hist') # creates a histogram using plotly
```

```python
df['A'].iplot(kind = 'hist', bins = 30)
```

```python
df.iplot(kind = 'hist')
```

```python
# lets say we want to create a plot for specific list of columns

df[['A','B']].iplot(kind = 'hist')
```

```python
df[['A','B']].iplot(kind = 'spread')
```

```python
df.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C', colorscale = 'rdylbu')

# essentially a scatter plot with an observations size based off of another column values
```
