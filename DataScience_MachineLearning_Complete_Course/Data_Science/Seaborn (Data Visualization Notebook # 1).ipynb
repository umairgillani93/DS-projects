### Seaborn Data Visualization:

```python
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # using built-in seaborn dataset for visualization purpose
```

```python
tips['total_bill'].head()
```

### distplot():

```python
sns.distplot(tips['total_bill'], bins = 30, kde = False)  # takes in a pandas series, 1-d numpy array, or python list and

# shows the distribution of the column. column name will be used to represent data axis
# in our case its 'total_bill'
```

```python
# and if we want to plot our data with 'kernal density estimation', we can simply go ahead and do the following

sns.distplot(tips['total_bill'], bins = 40, kde = True)
```

```python
tips.head()
```

```python
sns.distplot(tips['tip'], kde = True) # used to show the univariate graph
# univarite graph: graphs that consists of only one variable
```

```python
sns.distplot(tips['size'], kde = True)
```

### jointplot():

```python
# lets go ahead and create some more plots with seaborn

sns.jointplot(x = tips['total_bill'], y = tips['tip'], data = tips) # used to visuazlize bivariate data. Type of data in which one column is related to other and we have atlest 2 columns
```

```python
sns.jointplot(x = tips['size'], y = tips['tip'], data = tips, kind = 'hex')

# we can also adjust 'kind' argument to have a better visualization
```

```python
sns.jointplot(x = tips['total_bill'], y = tips['tip'], data = tips, kind = 'reg')

# shows the regressive plot
```

```python
sns.jointplot(x = tips['total_bill'], y = tips['tip'], data = tips, kind = 'kde')
```

### pairplot():

```python
# shows the plot for the numerical values of all the columns

sns.pairplot(data = tips)
```

```python
# we can alos pass in 'hue' value in order to categorize that specific column
# Remember! 'hue' is only used for categorical column and not numerical column

sns.pairplot(data = tips, hue = 'sex') # it will color the hue column based off of the categories
```

```python
# we can also adjust 'palette' argument in order to customize our colour scheme a bit more

sns.pairplot(data = tips, hue = 'sex', palette = 'coolwarm')
```

### rugplot():

```python
sns.rugplot(tips['total_bill']) # shows concentrated lines where values are higher
```

```python
# lets create an example with rugplot

import numpy as np

data = np.random.rand(25)
sns.rugplot(data)
```
