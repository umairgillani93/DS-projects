```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # module to visualize graphs with matplotlib
%matplotlib inline  # to see the plotted graphs in jupyter notebook
```

```python
x = np.linspace(0,5,10)
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y) # calling plot function off of class pyplot to visualize the graphs
```

```python
# similarly lets take another plotting variables

p = np.random.randint(1,10,5)
q = p * 2
```

```python
plt.plot(p,q) # reports back the plot on 'p' and 'q' axes

# if you are not using Jupyter Notebook, in order to show the graph you have to use the following method

plt.show()
```

```python
# we can also add labels and titles to our plot using matplotlib
# this is traditional matplotlib approach, in which we can directly call methods off of our plt variable
plt.plot(x,y)
plt.title('Title')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
```

```python
# we can also create subplots with matplotlib

plt.subplot(1,2,1)  # plots subplot with 1 row, 2 cols, and last '1' refers to the first figure
plt.plot(x, y, 'r')

plt.subplot(1,2,2) # plots subplot with 1 row, 2 cols, and last '2' refers to the second figure
plt.plot(y, x, 'b')
```

```python
# Creating figures, axes and drawing plot on those axes

fig = plt.figure(figsize = (10,6))

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig1 = plt.figure()  # in this approach we have a bit more control over our selections

# we can take our desired figure size as well as call differenct methods and attributes off of our created objects
# just like in OOP

axes = fig1.add_axes([0.2,0.2,0.9,0.9])

axes.plot(y,x)
axes.set_xlabel('x_axis')
axes.set_ylabel('y_axis')
axes.set_title('Title')
```

```python
# lets go ahead a create a plot inside a plot

fig2 = plt.figure()

axes1 = fig2.add_axes([0.2,0.2,0.9,0.9])
axes2 = fig2.add_axes([0.3,0.4,0.3,0.3])

axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')

axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')

axes1.set_title('Title')
axes2.set_title('Title')

axes1.plot(x,y)
axes2.plot(y,x)

# And that's how you can create plot inside a plot
```

```python
# creating subplots one more example

plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')

```
