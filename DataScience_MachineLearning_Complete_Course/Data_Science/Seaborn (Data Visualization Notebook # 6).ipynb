### Seaborn Regression plot:

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.rename(columns = {'sex':'gender'}, inplace = True)
tips.head()
```

### lmplot():

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips) # bivariate plot, takes in 2 numerical columns

# and shows a regressive plot
# a Regression plot is basically a scatter plot, with a linear fit on top of it
```

```python
# lets call hue argument, in order to have a better visualizatioin

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'gender')
```

```python
# lets add markers argument in our plot

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'gender', markers = ['x','v'])
```

```python
# instead of 'hue', we can also use 'col' argument to make seperation on the basis of categorical column

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'gender')
```

```python
# lets add argument 'row' as well and see the difference

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'gender', row = 'time')
```

```python
sns.lmplot(x = 'total_bill', y = 'size', col = 'smoker', row = 'gender', data = tips)
```

```python
# lets blend all the arguments together now

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, row = 'time', col = 'day', hue = 'gender')
```

```python
# we can also adjust 'aspect' and 'size' arguments as well
# aspect --> aspect ratio of each facet in inches

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'gender', col = 'day', aspect = 0.5,
          size = 7)
```
