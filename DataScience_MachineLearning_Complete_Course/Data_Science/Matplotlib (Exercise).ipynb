### Matplotlib Exercise:

```python
# lets grab some essential libraries and data first

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,100)
y = x * 2
z = x ** 2
```

```python
# if we are not using these commands, we can use plt.show() method in order to see the graphs
```

```python
figure = plt.figure() # creating a figure object
axes = figure.add_axes([0,0,1,1]) # adding axes to our figure
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
# now lets create two set of axes inside our figure

figure1 = plt.figure()

axes1 = figure1.add_axes([0,0,1,1])
axes2 = figure1.add_axes([0.2,0.5,.2,.2])

axes1.plot(x,y)
axes2.plot(x,y)
```

```python
axes1.plot(x,y)
```

```python
# lets do another example

figure2 = plt.figure()

axes1 = figure2.add_axes([0,0,1,1])
axes1.set_xlabel('X axis')
axes1.set_ylabel('Z axis')
axes1.set_title('X-Z PLOT')
axes2 = figure2.add_axes([0.2,0.5,.4,.4])
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('X-Y PLOT')
axes1.plot(x,z, color = 'red')  # plot x,z on axes1 with 'red' color
axes2.plot(x,y, color = 'blue') # plot x,y on axes2 with 'blue' color
```

### Creating Subplots:

```python
# lets now go ahead and create some subplots
# and plot our axes on them

# plt.subplots(nrows = 1, ncols = 2)
# plt.plot(x,y, color = 'blue', linestyle = '--')
# plt.plot(x,z, color = 'red')

fig,ax = plt.subplots(nrows = 1, ncols = 2, figsize = (8,3))

ax[0].plot(x,y,color = 'red', linestyle = '--')
ax[1].plot(x,z, color = 'blue')
```
