### PairGrid():

#### shows subplots without plotting the graphs, and gives us a little bit more control

```python
sns.PairGrid(data = iris)
```

```python
# let's go ahead and create some plots in these subplots

g = sns.PairGrid(data = iris)
g.map_diag(sns.distplot) # plots the distplot in diagonal
g.map_upper(plt.scatter) # plots the scatter plot in upper region of diagonal
g.map_lower(sns.kdeplot) # plots kdeplot in lower region of diagonal
```

```python
h = sns.PairGrid(data = iris)
h.map_diag(sns.distplot)
h.map_upper(plt.scatter)
h.map_lower(sns.kdeplot)
```

### FacetGrid():

```python
# for FacetGrid lets call our 'tips' dataframe from seaborn

tips = sns.load_dataset('tips')
tips.rename(columns = {'sex':'gender'}, inplace = True) # just changed the name of column to 'gender'
tips.head()
```

```python
# FacetGrid also creates subplots for us, in which we can create the plots we want

g = sns.FacetGrid(data = tips, row = 'time', col = 'gender')
g.map(sns.distplot, 'total_bill')

# FacetGrid draws 'time' VS 'gender' subplots
# and then fill those up with third parameter 'total_bill' with chosen plot type
```

```python
h = sns.FacetGrid(data = tips, row = 'gender', col = 'smoker')
h.map(sns.distplot, 'tip')
```

```python
tips.head()
```

```python
i = sns.FacetGrid(data = tips, row = 'gender', col = 'day') # Categorical values
i.map(sns.distplot, 'size') # numerical value
```

```python
k = sns.FacetGrid(data = tips, row = 'gender', col = 'time')
k.map(plt.scatter, 'total_bill', 'tip') # creating plots with plt.scatter
```
