## Subplots:

```python
# lets create subplots now using OOP method

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,5,10)
y = x ** 2
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) # creates the subplots of a figure
```

```python
# and if we were to create more subplots off of our figure
# just pass in nrows and ncols values

fig,axes = plt.subplots(nrows = 3, ncols = 3)

plt.tight_layout() # fixes the overlaped values of subplots
```

```python
axes # axes is a numpy array having corresponding values for each figure
```

```python
# since axes is an array of figures, we can iterate through it
fig,axes = plt.subplots(nrows = 1, ncols = 2)

for axis in axes:
    axis.plot(x,y)
    
# its interesting to notice that when we use OOP methods to create subplots, it takes multiples 
# axes as as array, through which we can iterate over
```

### Indexing Axes:

```python
# since axes is an iterable we can also call indexes off of it
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('First Axes')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')

axes[1].plot(y,x)
axes[1].set_title('Second Axes')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
```

### Figure Size and DPI:

```python
# we can also set figure size with matplotlib

fig = plt.figure(figsize = (3,2), dpi = 100)
# figsize = 3 * 2 inches, dpi --> dots per inch

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,3))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# lets go ahead and create one last example of subplots

p = np.random.randint(1,100,10)
q = np.sqrt(p)
```

### Saving file with Matplotlib:

```python
# we can also save a file with Matplotlib

a = np.arange(1,10)
b = np.sin(a)
```

```python
a
```

```python
b
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,3))

for axis in axes:
    axis.plot(a,b)
    axis.plot(b,a)
    
plt.savefig('my_fig.png', dpi = 200) # saves the figure with figure name and dpi's
```

```python
figure = plt.figure(figsize = (8,3))

axes = figure.add_axes([0.2,0.2,0.8,0.8])
axes.plot(a,b)
axes.plot(b,a)  # both plots are plotted on a single set of axes
```

### legend():

```python
# legend method helps us discriminate multiple plots within same set of axes

a = np.arange(1,10)

figure = plt.figure()
axes = figure.add_axes([0,0,1,1])

axes.plot(a, a ** 0.5, label = 'A Squared')
axes.plot(a, a ** 0.25, label = 'A Cubed')

axes.legend() # creates colour tags to associate corresponding graphs
```
