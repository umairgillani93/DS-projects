### K Means Clustering:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.5, random_state = 101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
```

### Creating Clusters:

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 4)
```

```python
k_means.fit(data[0])
```

```python
k_means.cluster_centers_
```
